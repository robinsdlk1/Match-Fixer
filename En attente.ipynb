```python
files = ['22-23.csv', '13-14.csv', '14-15.csv', '15-16.csv', '16-17.csv', '17-18.csv', '18-19.csv', '19-20.csv', '20-21.csv', '21-22.csv']
dir_path = 'Odds data L1\\'
dfs = [pd.read_csv(dir_path + file) for file in files]
df = pd.concat(dfs, ignore_index=True)

#On veut trier par date.
df['Date']

def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%d/%m/%y')
    except ValueError:
        return pd.to_datetime(date_str, format='%d/%m/%Y')
    
df['Date'] = df['Date'].apply(parse_date)
df.sort_values('Date', inplace=True)
df
```
